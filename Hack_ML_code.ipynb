```python

import json_lines
import jsonlines
import os, json
import pandas as pd
import re
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
path_to_json = 'E:\data'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.jsonl')]
print(json_files[9:33])  # for me this prints ['foo.json']
```

################ Code Starts Here ###########################

```python
dfs=[]
```

```python
for f in json_files[0:8]:
    fname= "E:\\Data\\"+ f
    jsonObj = pd.read_json(fname,lines = True)
    dfs.append(jsonObj)
    print(f)
    print("\n")
```

```python
temp = pd.concat(dfs, ignore_index=True)
temp.count()
```

```python
df=temp
```

```python
#cleaning the data
df['cleaned_tweet'] = df['text'].replace("  ", " ")

#remove user mentions

df['cleaned_tweet'] = df['cleaned_tweet'].replace(r'^(@\w+)',"", regex=True) 
#remove 'rt' in the beginning

df['cleaned_tweet'] = df['cleaned_tweet'].replace(r'^(RT @)',"", regex=True)

#remove_URL(x):

df['cleaned_tweet'] = df['cleaned_tweet'].replace(r'https.*$', "", regex = True)

#remove 'amp' in the text

df['cleaned_tweet'] = df['cleaned_tweet'].replace(r'amp',"", regex = True)

df['cleaned_tweet'] = df['cleaned_tweet'].replace(r'https',"", regex = True)

#remove extra spaces in the tweet

df['cleaned_tweet'] = df['cleaned_tweet'].replace(r'^\s+|\s+$'," ", regex=True)
```

```python
df = df.loc[df['lang'] == 'en']
```

```python
df['cleaned_tweet']
```

```python
text=df['cleaned_tweet']
text_l=text.str.lower()
```

```python
t=[]
p=0
for str2 in text_l:
    p=p + str2.count('work from home')
    if(str2.count('work from home')>0):
        t.append(str2)
print(p)
```

```python
for sentence in t:
 
    sid = SentimentIntensityAnalyzer()
    print(sentence)
    ss = sid.polarity_scores(sentence)
    for k in sorted(ss):  
        print('{0}: {1}, '.format(k, ss[k]), end='')
        print()
    print()
    
```

```python
from textblob import TextBlob
for sentence in t:
    blob = TextBlob(sentence)
    print(sentence)
    print(blob.sentiment)
    print()

```
